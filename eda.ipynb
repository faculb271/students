```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Students_Grading_Dataset.csv") # Lo primero que hago es examinar el Dataframe, en este caso para ver que contiene. 
df

# Aqui ya podemos observar que hay cosas que no nos sirven en la columnas. Como el nombre, el apellido y el mail.
```

```python
df = df.drop(columns=["First_Name","Last_Name","Email"]) #Eliminamos las columnas que no no sirven
df.isnull().sum() # Con esto podemos ver los valores nulos del dataframe, de paso puedo ver el nombre de las columnas

# Aqui podemos ver que, hay muchos datos de Attendace faltantes, por lo tanto podemos imputarlo.
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data = df,x="Department", y="Study_Hours_per_Week",hue="Family_Income_Level")

plt.xticks(rotation=45)
plt.title("Attendance por Department")
plt.show()
```

```python
df["Attendance (%)"] = df["Attendance (%)"].fillna(df.groupby(["Department","Gender","Age"])["Attendance (%)"].transform("mean")) 
# Reemplazamos en base a caracteriticas Similares
df.isnull().sum() 
```
